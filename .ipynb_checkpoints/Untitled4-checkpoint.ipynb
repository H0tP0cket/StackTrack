```python
import requests
from bs4 import BeautifulSoup

def extract_university_names(url):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all <td> tags containing <a> tags (university names)
        university_list = soup.find_all('td')
        
        # Extract the text from the <a> tags and store it in a list
        universities = [td.find('a').get_text().strip() for td in university_list if td.find('a')]
        
        return universities
    else:
        # Print an error message if the request was not successful
        print("Failed to retrieve content from the URL")
        return None

# URL of the website containing the list of universities
url = "https://www.4icu.org/us/a-z/"

# Call the function to extract university names from the website
university_names = extract_university_names(url)

# Print the extracted university names
if university_names:
    for i, name in enumerate(university_names, start=1):
        print(f"{i}. {name}")
```

```python
def classify(n):
    total, k = 0, 1
    while k <= n:
        if n // (k + 1) == 0:
            total += (k + 1)
        k = k + 1
    if total == n:
        return 'perfect'
    elif total < n:
        return 'deficient'
    else:
        return 'abundant'

def main():
    # Test cases
    print(f"Classify 6: {classify(6)}")    # perfect
    print(f"Classify 10: {classify(10)}")  # deficient
    print(f"Classify 12: {classify(12)}")  # abundant

    # You can also ask for user input to test other numbers
    n = int(input("Enter a number to classify: "))
    print(f"Classify {n}: {classify(n)}")

# Run the main function
if __name__ == "__main__":
    main()
```
